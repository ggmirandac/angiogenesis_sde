```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os 
```

```python
lists_personal_comp = os.listdir('personal_comp')
linear = {}
exponential = {}
constnat = {}
for file in lists_personal_comp:
    separated = file.split('_')
    gradient = separated[0]
    H = (separated[-2] + '_' + separated[-1].split('.')[0]).replace('_','.')
    if gradient == 'linear':
        linear[H] = pd.read_csv('personal_comp/' + file)
    elif gradient == 'exponential':
        exponential[H] = pd.read_csv('personal_comp/' + file)
    elif gradient == 'constant':
        constnat[H] = pd.read_csv('personal_comp/' + file)
```

```python
fig, ax = plt.subplots(3, 1, figsize=(10,10))
for H in linear:
    sns.histplot(linear[H], label = H, ax = ax[0], alpha = 0.1,fill = True, 
                 line_kws={'linewidth': 0, 'alpha': 0.1}, multiple='dodge', element='step')
ax[0].set_title('Linear')

for H in exponential:
    sns.kdeplot(exponential[H], label = H, ax = ax[1])
ax[1].set_title('Exponential')

for H in constnat:
    sns.kdeplot(constnat[H], label = H, ax = ax[2])
ax[2].set_title('Constant')
```
